```python
def get_instruction(position: int, data: list) -> tuple:
    instruction = data[position]
    op = instruction % 100
    mode1 = instruction // 100 % 10
    mode2 = instruction // 1000 % 10
    # Thanks to Karl Uibo for the instruction breakdown above
    if op < 3: 
        param_count = 3
    elif op < 5: 
        param_count = 1
    elif op < 7:
        param_count = 2
    elif op < 9:
        param_count = 3
    else: 
        param_count = 0
    return op, [mode1, mode2], param_count

def get_parameters(position: int, count: int, data: list) -> list:
    parameters = data[position:position + count]
    return parameters

def get_values(params: list, param_modes: list, data: list) -> list:
    values = []
    for p, m in zip(params, param_modes):
        if m == 1:
            values.append(p)
        else:
            values.append(data[p])
    return values

def intcode_computer(data: list, opinput: int):
    # get phase setting (0-4)
    instruction_set = data[:]
    position = 0
    opoutput = []
    op, param_modes, param_count = get_instruction(position, instruction_set)
    
    print('Position | Op Code | Parameter Modes | Parameter Count | Parameters        | Values')
    
    while True:
        skip_normal_move = False
        parameters = get_parameters(position+1, param_count, instruction_set)
        values = get_values(parameters, param_modes, instruction_set)
    
        print(f'{position:<8} | {op:^7} | {str(param_modes):^15} | {param_count:^15} | {str(parameters):<17} | {str(values):<10}')
    
        if op == 1:      # Add: 3 parameters
            instruction_set[parameters[2]] = values[0] + values[1]
        elif op == 2:    # Multiply: 3 parameters
            instruction_set[parameters[2]] = values[0] * values[1]
        elif op == 3:    # Save Input: 1 parameter
            instruction_set[parameters[0]] = opinput
        elif op == 4:    # Output Value: 1 parameter
            opoutput.append(values[0])
        
        
        # START TROUBLE
        elif op == 5:    # Jump if True: 2 parameters
            if parameters[0] != 0:
                position = values[1]
                skip_normal_move = True
        elif op == 6:    # Jump if False: 2 parameters
            if parameters[0] == 0:
                position = values[1]
                skip_normal_move = True
        elif op == 7:    # Less Than: 3 parameters
            if parameters[0] < parameters[1]:
                instruction_set[parameters[2]] = 1
            else:
                instruction_set[parameters[2]] = 0
        elif op == 8:    # Equals: 3 parameters
            if parameters[0] == parameters[1]:
                instruction_set[parameters[2]] = 1
            else:
                instruction_set[parameters[2]] = 0
        # END TROUBLE
        
        elif op == 99:
            return op, opoutput, instruction_set
        else:
            print(f'Invalid Opcode: {op}')
            return
        if not skip_normal_move:
            position += param_count + 1
        op, param_modes, param_count = get_instruction(position, instruction_set)
```
